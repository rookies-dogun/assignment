```python
import os
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import json

load_dotenv()



def search_books(query, options={"price":0,"publisher":"", "save": False}):
    
    # 네이버 요청 기본설정
    naver_api_url = "https://openapi.naver.com/v1/search/book.json"

    naver_request_headers = {
        "X-Naver-Client-Id": os.getenv("NAVER_CLIENT_ID"),
        "X-Naver-Client-Secret": os.getenv("NAVER_CLIENT_SECRET"),
    }
    
    res = requests.get(naver_api_url, headers=naver_request_headers, params=query)

    if res.ok:
        # 첫부분 출력
        
        print("===",end="")   
        # 책 위치
        book_list = res.json()["items"]
        
        # 가격 데이터 정제제
        if "price" in options and options["price"] >= 0:
            print(f" 가격이 {options["price"]} 이상",end="")   
            for book in book_list[:]:
                if int(book["discount"]) < options["price"]:
                    book_list.remove(book)
        
        # 출판사 데이터 정제
        if "publisher" in options and len(options["publisher"]) > 0:
            print(",", end="") if "price" in options and options["price"] >= 0 else ""
            print(f' {options["publisher"]} 출판판',end="")  
            for book in book_list[:]:
                if book["publisher"] != options["publisher"]:
                    book_list.remove(book)
                  
        # 마지막 출력력
        print(" 도서 목록 ===")            
                    
        for idx, book in enumerate(book_list):
            print(f"{idx + 1}. {book["title"]}")
            print(f"\t저자 : {book["author"]}")    
            print(f"\t출판사 : {book["publisher"]}")    
            print(f"\t출판일 : {book["pubdate"][:4]}-{book["pubdate"][4:6]}-{book["pubdate"][6:]}")    
            print(f"\t가격 : {book["discount"]}")            
            print(f"\t설명 : {book["description"][:100]}\n")
            
        # 저장
        if "save" in options and options["save"]:
            
            os.makedirs("./data", exist_ok=True) 
            with open("./data/books.json", "w") as json_file:
                json.dump(book_list, json_file, indent=4, ensure_ascii=False)
                
            
            
        
    else:
        print(f'Error Code : {res.status_code}')
        print(res.text)
        

```

```python

query = {
    "query": "주식",
    "display": 10,
    "start": 1
}   


# 1. 질문 :  검색어로  찾은  책 목록 출력하기
search_books(query)
# 2. 질문 :  검색어로  찾은  책 목록 중에서 가격이 2만원 이상인 책만 출력하기
search_books(query, options={"price": 20000})
# 3. 질문 :  검색어로  찾은  책 목록 중에서 출판사가 인피니티북스인 책만 출력하기
search_books(query, options={"publisher": "가이아의어깨"})
# 4. 질문 :  검색어로  찾은  책 목록을 json 파일로 저장하기
search_books(query, options={"save": True})

```

```python
import os
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import json

load_dotenv()


def search_shops(query, options={"price":0,"publisher":"", "save": False}):
    
    # 네이버 요청 기본설정
    naver_api_url = "https://openapi.naver.com/v1/search/shop.json"

    naver_request_headers = {
        "X-Naver-Client-Id": os.getenv("NAVER_CLIENT_ID"),
        "X-Naver-Client-Secret": os.getenv("NAVER_CLIENT_SECRET"),
    }
    
    res = requests.get(naver_api_url, headers=naver_request_headers, params=query)

    if res.ok:
        
        # # 책 위치
        shop_list = res.json()["items"]
        
        # # 가격 데이터 정제제
        if "price" in options and options["price"] > 0:
            print(f"=== {options["price"]}원 이하 상품 ===")
            for shop in shop_list[:]:
                if int(shop["lprice"]) > options["price"]:
                    shop_list.remove(shop)
        
        # # 출판사 데이터 정제
        elif "mallName" in options and len(options["mallName"]) > 0:
            print(f'=== {options["mallName"]} 상품 ===')  
            for shop in shop_list[:]:
                if shop["mallName"] != options["mallName"]:
                    shop_list.remove(shop)
                  
        # # 마지막 출력
        else:
            print("=== 쇼핑 검색 결과 ===")            
                    
        for idx, shop in enumerate(shop_list):
            print(f"{idx + 1}. {shop["title"]}")
            print(f"\t가격격 : {shop["lprice"]}")    
            print(f"\t브랜드 : {shop["brand"]}")    
            print(f"\t쇼핑몰 : {shop["mallName"]}")            
            print(f"\t링크크 : {shop["link"]}\n")
            
        # # 저장
        # if "save" in options and options["save"]:
            
        #     os.makedirs("./data", exist_ok=True) 
        #     with open("./data/books.json", "w") as json_file:
        #         json.dump(book_list, json_file, indent=4, ensure_ascii=False)
            
    else:
        print(f'Error Code : {res.status_code}')
        print(res.text)
```

```python
payload = {
    'query':  '가디건건', #'가디건',
    'display': 100,
    'sort': 'sim'
    }

# 1. 질문 :  검색어로  찾은  Shop의 상품  목록 출력하기
search_shops(payload)
# 2. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 가격이 50,000원 이하인 상품만 출력하기
search_shops(payload, options={"price":20000})
# 3. 질문 :  검색어로  찾은  Shop의 상품  목록 중에서 특정 쇼핑몰 상품만 출력하기
search_shops(payload, options={"mallName":"네이버"})
```
